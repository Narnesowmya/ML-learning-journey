```python
#Fibonacci Series means the next number is the sum of previous two numbers
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b
    print()
fibonacci(15)
```

```python
#Factorial of a Number
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

print(factorial(5))
```

```python
#Prime Number Check
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

print(is_prime(7))
print(is_prime(10))
```

```python
#Print Prime Numbers up to N
def print_primes(n):
    for num in range(2, n + 1):
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num)

print_primes(100)
```

```python
#Reverse a Number means to reverse the digits of a number
def reverse_number(n):
    rev = 0
    while n > 0:
        rev = rev * 10 + n % 10
        n //= 10
    return rev

print(reverse_number(1234))
```

```python
#Sum of Digits means to caliculate the sum of all digits in a number
def sum_of_digits(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total

print(sum_of_digits(123))
```

```python
#Palindrome Number means the number which is same when read from left to right and right to left
def is_palindrome(n):
    original = n
    rev = 0

    while n > 0:
        rev = rev * 10 + n % 10
        n //= 10

    return original == rev

print(is_palindrome(121))
print(is_palindrome(123))
```

```python
#Armstrong Number means the sum of cubes of each digit is equal to the number itself
def is_armstrong(n):
    temp = n
    total = 0

    while temp > 0:
        digit = temp % 10
        total += digit ** 3
        temp //= 10

    return total == n

print(is_armstrong(153))
print(is_armstrong(123))
```
