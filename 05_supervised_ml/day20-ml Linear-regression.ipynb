```python
#ml -Supervisied Learning Basics, Regression .
#Regression is a type of supervised machine learning used to predict continuous numerical values based on input features.
```

```python
#Regression Types:
#1. Linear Regression: Models the relationship between a dependent variable and one or more independent variables
#   using a linear equation. It's commonly used for predicting values like house prices, sales, etc.
#2. Polynomial Regression: Extends linear regression by allowing for non-linear relationships between the dependent
#   and independent variables. It fits a polynomial equation to the data.
#3. Ridge Regression: A type of linear regression that includes a regularization term to prevent overfitting by
#   penalizing large coefficients.
#4. Lasso Regression: Similar to ridge regression but uses L1 regularization, which can shrink some coefficients
#   to zero, effectively performing feature selection.
#5. Logistic Regression: Used for binary classification problems, predicting the probability of a categorical
#   outcome based on input features.
#6. Support Vector Regression (SVR): An extension of support vector machines for regression tasks, which aims to
#   find a function that deviates from the actual observed values by a value no greater than a specified margin.
#7. Decision Tree Regression: Uses decision trees to model the relationship between input features and continuous
#   target variables by splitting the data into subsets based on feature values.
#8. Random Forest Regression: An ensemble method that combines multiple decision trees to improve prediction
#   accuracy and reduce overfitting.
#9. Gradient Boosting Regression: An ensemble technique that builds models sequentially, where each new model
#   corrects the errors of the previous ones, often leading to high predictive performance.
#10. Neural Network Regression: Utilizes artificial neural networks to model complex relationships between input
#    features and continuous target variables, capable of capturing non-linear patterns in data.
#11. Bayesian Regression: A probabilistic approach to regression that incorporates prior distributions on the
#    model parameters and updates these beliefs based on observed data.
#12. Elastic Net Regression: Combines both L1 and L2 regularization techniques to improve model performance
#    and handle multicollinearity among features.
```

```python
#y = mx + b
"""x = input 

y = output 

m = slope (rate of change)

b = intercept (base value)"""
```

```python
#linear Regression  Predicts output using a straight-line relationship.
# Its answers that if x changes ,how much does y chanege?
```

```python
"""Training does NOT mean memorizing data.

Training means:

Finding the best values of m and b
such that the line fits the data as closely as possible."""
```

```python
"""We compare:

Actual value (real y)

Predicted value (model’s ŷ)

Error (residual):
error = actual y - predicted ŷ
	​


If prediction is perfect → error = 0
If not → error exists"""
```

```python
"""One error is not enough. Loss function
We have many data points.

So we measure overall error using a loss function.

Most common:

Mean Squared Error (MSE)
MSE = (1/n) * Σ(actual y - predicted ŷ)²
      
"""

"""

Why square?

No negative values

Penalizes big mistakes more"""
```

```python
"""THE GOAL OF LINEAR REGRESSION?

Very simple:

Choose m and b that minimize the loss (MSE)

That’s it.

This process is called:

Optimization"""
```

```python
"""## Linear Regression – Understanding

Linear regression is a supervised learning algorithm used to predict
continuous values. It models the relationship between input features
and the target variable using a linear equation. The model learns
parameters that minimize prediction error on training data and uses
the learned equation to make predictions on new data.
"""
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split #sklean means its a library for machine learning used for splitting data into training and testing sets.
from sklearn.linear_model import LinearRegression #Linear regression model from scikit-learn
from sklearn.metrics import mean_squared_error, r2_score #metrics to evaluate model performance

import matplotlib.pyplot as plt # used for plotting graphs and visualizations.
```

```python
#We’ll use a simple synthetic dataset first.
# Simple dataset
hours = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
scores = np.array([35, 40, 50, 55, 65, 70, 78, 85])
X = hours
y = scores

```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.coef_
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
pd.DataFrame({
    "Actual": y_test,
    "Predicted": y_pred
})
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
plt.scatter(X, y, color="blue", label="Actual data")
plt.plot(X, model.predict(X), color="red", label="Regression line")
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.legend()
plt.show()
```

```python
"""## Model Interpretation

The linear regression model learned a positive relationship between
study hours and exam score. The slope indicates that as study hours
increase, the predicted score increases. The R² score shows how well
the model explains the variance in the data.
"""
```
