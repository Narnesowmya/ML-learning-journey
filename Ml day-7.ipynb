```python
# Day - 7  pandas 
import pandas as pd   # to read csv file

df = pd.read_csv(r"C:\Users\USER\Downloads\Titanic-Dataset.csv") #r"" tells Python:“Do NOT treat backslashes as escapes”
 # reading csv file using pandas were i have downloaded the titanic dataset from kaggle and provided the path of dataset .
print(df.head())  # printing first 5 rows of dataset
print(df.tail())  # printing last 5 rows of dataset
print(df.info())  # printing info of dataset
```

```python
df.isnull().sum() # checking for null values in dataset
```

```python
df["Age"].median()
df["Age"].fillna(df["Age"].median(), inplace=True)
# filling null values in age column with median value of age column.
```

```python
df[df["Age"] > 30]
# filtering dataset where age is greater than 30 years
```

```python
df.groupby("Sex")["Survived"].mean()
# grouping dataset based on sex and finding mean of survived column    
```

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
# creating a new column family size by adding sibsp , parch and 1 (1 for self)  
```

```python
df.groupby("FamilySize")["Survived"].mean()
# grouping dataset based on familysize and finding mean of survived column
```
