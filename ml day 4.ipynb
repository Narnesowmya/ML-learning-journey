```python
#Data Structures in python used for Machine Learning
#Lists, Tuples, Sets, Dictionaries
#lists is a collection which is ordered and changeable . Allows duplicate members.
#list is defined using square brackets []
#elements in a list can be of different data types
#elements can be accessed using indexing and slicing ,indexing starts from 0
#elements can be added using append(), insert(),extend(), removed using remove(),pop(),clear()
```

```python
# list basic Examples 
mylist = [1,2,3,4,5]
print(mylist)
print(type(mylist))
print(mylist[0]) # accessing first element
print(mylist[1:4]) #slicing elements from index 1 to 3
mylist.append(6) # adding element at the end
print(mylist)
mylist.insert(0,0) # adding element at index 0
print(mylist)
mylist.remove(3) # removing element with value 3
print(mylist)
mylist.pop() # removing last element
print(mylist)
mylist.clear() # removing all elements
print(mylist)   
```

```python
#Python Program to Find Second Largest Number in a List
def second_largest(numbers): #function to find second largest number
    first = second = float('-inf')#initialize first and second largest to negative infinity
    for number in numbers:#iterate through the list
        if number > first:#check if current number is greater than first largest
            second = first#update second largest
            first = number#update first largest
        elif first > number > second:#check if current number is between first and second largest
            second = number#update second largest
    return second#return second largest number
numbers = [10, 20, 4, 45, 99] 
print("Second largest number is:", second_largest(numbers)) 
```

```python
#Remove Duplicates from a sorted array of list
def remove_duplicates(arr): #function to remove duplicates  
    if not arr: #check if array is empty
        return 0
    write_index = 1 #initialize write index
    for i in range(1, len(arr)): #iterate through the array starting from index 1
        if arr[i] != arr[i - 1]: #check if current element is different from previous element
            arr[write_index] = arr[i] #update the array at write index
            write_index += 1 #increment write index
    return write_index #return the length of array after removing duplicates
arr = [1, 1, 2, 2, 3, 4, 4, 5]
new_length = remove_duplicates(arr)
print("Array after removing duplicates:", arr[:new_length])
print("New length of array:", new_length)
```

```python
#2.Tuples
#Tuples is a collection which is ordered and unchangeable . Allows duplicate members.
#tuple is defined using parentheses ()
#elements in a tuple can be of different data types
#elements can be accessed using indexing and slicing ,indexing starts from 0
#elements cannot be added or removed as tuples are immutable
# tuple basic Examples
mytuple = (1,2,3,4,5)
print(mytuple)
print(type(mytuple))
print(mytuple[0]) # accessing first element
print(mytuple[1:4]) #slicing elements from index 1 to 3
#mytuple.append(6) # adding element at the end - will raise error
#print(mytuple)
#mytuple.remove(3) # removing element with value 3 - will raise error
#print(mytuple)
#mytuple.pop() # removing last element - will raise error
#print(mytuple)
#mytuple.clear() # removing all elements - will raise error
#print(mytuple)
#Python Program to Find the Length of a Tuple
def tuple_length(tup): #function to find length of tuple
    count = 0 #initialize count
    for _ in tup: #iterate through the tuple
        count += 1 #increment count
    return count #return length of tuple
mytuple = (10, 20, 30, 40, 50)
print("Length of the tuple is:", tuple_length(mytuple))
```

```python
#Remove Empty Tuples from a List 
def remove_empty_tuples(tup_list): #function to remove empty tuples
    return [tup for tup in tup_list if tup] #list comprehension to filter out empty tuples
tup_list = [(1, 2), (), (3, 4), (), (5, 6)]
result = remove_empty_tuples(tup_list)
print("List after removing empty tuples:", result)
```

```python
#Python | Sort Tuples in Increasing Order by any key
def sort_tuples(tup_list, index): #function to sort tuples by given index
    return sorted(tup_list, key=lambda x: x[index])#sort using lambda function as key
tup_list = [(1, 3), (2, 1), (4, 2)]
sorted_by_second = sort_tuples(tup_list, 1)
print("Tuples sorted by second element:", sorted_by_second)
```

```python
#Sets
#Sets is a collection which is unordered and unindexed . No duplicate members allowed.
#set is defined using curly braces {}
#elements in a set can be of different data types
#elements cannot be accessed using indexing and slicing as sets are unordered
#elements can be added using add(), update(), removed using remove(), discard(), pop(),clear()
# set basic Examples
myset = {1,2,3,4,5}
print(myset)
print(type(myset))
print(3 in myset) # check if 3 is in set
myset.add(6) # adding element
print(myset)
myset.remove(3) # removing element
print(myset)
myset.pop() # removing arbitrary element
print(myset)
myset.clear() # removing all elements
print(myset)
#Python Program to Find the Length of a Set
def set_length(s): #function to find length of set
    count = 0 #initialize count
    for _ in s: #iterate through the set
        count += 1 #increment count
    return count #return length of set
myset = {10, 20, 30, 40, 50}
print("Length of the set is:", set_length(myset))
#Python Program to Remove Duplicates from a List using Set
def remove_duplicates_from_list(lst): #function to remove duplicates using set
    return list(set(lst)) #convert list to set and back to list
mylist = [1, 2, 2, 3, 4, 4, 5]
new_list = remove_duplicates_from_list(mylist)
print("List after removing duplicates:", new_list)
```

```python
#Longest Consecutive Subsequence  using sets
def longest_consecutive_subsequence(arr): #function to find longest consecutive subsequence
    num_set = set(arr) #convert list to set for O(1) lookups
    longest_streak = 0 #initialize longest streak
    for num in num_set: #iterate through the set
        if num - 1 not in num_set: #check if it's the start of a sequence
            current_num = num #initialize current number
            current_streak = 1 #initialize current streak
            while current_num + 1 in num_set: #check for consecutive numbers
                current_num += 1 #increment current number
                current_streak += 1 #increment current streak
            longest_streak = max(longest_streak, current_streak) #update longest streak
    return longest_streak #return longest consecutive subsequence length
arr = [100, 4, 200, 1, 3, 2]
print("Length of longest consecutive subsequence is:", longest_consecutive_subsequence(arr))
```

```python
#Dictionaries
#Dictionaries is a collection which is unordered, changeable and indexed . No duplicate members allowed.
#dictionary is defined using curly braces {} with key-value pairs 
# dict basic Examples
mydict = {'a':1, 'b':2, 'c':3}
print(mydict)
print(type(mydict))
print(mydict['a']) # accessing value for key 'a'
print(mydict.keys()) # accessing all keys
print(mydict.values()) # accessing all values
mydict['d'] = 4 # adding key-value pair
print(mydict)
del mydict['b'] # removing key-value pair
print(mydict)
mydict.pop('c') # removing key-value pair using pop
print(mydict)
mydict.clear() # removing all key-value pairs
print(mydict)

#Python Program to Find the Length of a Dictionary
def dict_length(d): #function to find length of dictionary
    count = 0 #initialize count
    for _ in d: #iterate through the dictionary
        count += 1 #increment count
    return count #return length of dictionary
mydict = {'a': 10, 'b': 20, 'c': 30}
print("Length of the dictionary is:", dict_length(mydict))
#Python Program to Merge Two Dictionaries
def merge_dictionaries(dict1, dict2): #function to merge two dictionaries
    merged_dict = dict1.copy() #copy first dictionary
    merged_dict.update(dict2) #update with second dictionary
    return merged_dict #return merged dictionary
dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}
merged = merge_dictionaries(dict1, dict2)
print("Merged dictionary:", merged)
```
