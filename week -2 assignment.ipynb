```python
import pandas as pd

df = pd.read_csv(r"C:\Users\USER\Downloads\Titanic-Dataset.csv") # loading the dataset .
df.head() # displaying the first 5 rows of the dataset.
```

```python
mean_age = df["Age"].mean()  # calculating the mean of the "Age" column. mean 
mean_fare = df["Fare"].mean() # calculating the mean of the "Fare" column. fare mean

mean_age, mean_fare
```

```python
var_age = df["Age"].var() # calculating the variance of the "Age" column. 
std_age = df["Age"].std() # calculating the standard deviation of the "Age" column.

var_fare = df["Fare"].var()
std_fare = df["Fare"].std()

var_age, std_age, var_fare, std_fare
```

```python
df[["Age", "Fare"]].corr() # calculating the correlation between "Age" and "Fare" columns.
# The correlation matrix shows the relationship between the two variables.
#The correlation between Age and Fare is weak, indicating that age does not strongly influence ticket price.
```

```python
df[["Survived", "Age", "Fare"]].corr() # correlation matrix for Survived, Age, and Fare columns.
#Fare shows a stronger relationship with survival than age.
```

```python
import matplotlib.pyplot as plt # importing matplotlib for visualization.

plt.hist(df["Age"].dropna(), bins=20) # plotting a histogram for the "Age" column, dropping NaN values.
plt.xlabel("Age") # label for x-axis.
plt.ylabel("Count") # label for y-axis.
plt.title("Age Distribution") # title of the plot.
plt.show() # displaying the plot.
```

```python
plt.hist(df["Fare"], bins=30) # histogram to visualize distribution of fare column in titanic dataset
plt.xlabel("Fare") # x-axis label
plt.ylabel("Count")# y-axis label
plt.title("Fare Distribution")# title of the plot
plt.show()
```

```python
plt.boxplot(df["Fare"]) # boxplot to visualize the distribution of fare column in titanic dataset
plt.title("Fare Boxplot") # title of the plot
plt.show()
```

```python
survival_prob = df["Survived"].mean() # calculating the probability of survival in the dataset.
survival_prob # displaying the survival probability.
```
