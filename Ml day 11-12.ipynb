```python
#correlation measures the strength and direction of a linear relationship between two variables
#values range from -1 to 1  
#-1 indicates a perfect negative linear relationship
#1 indicates a perfect positive linear relationship
#0 indicates no linear relationship
```

```python
# if Height ↑ → Weight ↑ (positive)

# if Speed ↑ → Travel time ↓ (negative)

# if Shoe size vs marks → no relation . Its called  correlation .
```

```python
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

np.corrcoef(x, y) # corrcoef function gives correlation coefficient matrix . what happens is it gives correlation of x with x , x with y , y with x , y with y in matrix form . so diagonal elements will always be 1 because its correlation of same arrays .

```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 8, 6, 4, 2])

np.corrcoef(x, y)
#negative correlation because as x increases y decreases  expalin this with real life example
# if study hours increase → marks decrease (negative correlation)
# if temperature increase → hot chocolate sales decrease (negative correlation)
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([7, 3, 9, 2, 6])

np.corrcoef(x, y)
#no correlation because there is no relation between x and y
# if shoe size vs marks → no relation (no correlation)
# if age vs favorite color → no relation (no correlation)
```
