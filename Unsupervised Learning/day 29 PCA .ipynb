```python
#PCA — PRINCIPAL COMPONENT ANALYSIS means it is a dimensionality reduction technique used in unsupervised learning to transform high-dimensional data into a lower-dimensional space while preserving as much variance as possible. PCA works by identifying the directions (principal components) in which the data varies the most and projecting the data onto these directions. The first principal component captures the most variance, followed by the second, and so on. By selecting a subset of the principal components, we can reduce the dimensionality of the data while retaining important information. PCA is commonly used for visualization, noise reduction, and feature extraction in various applications such as image processing, gene expression analysis, and customer segmentation.
```

```python
"""PCA IS:

A dimensionality reduction technique
that compresses data while preserving maximum information"""
```

```python
"""Use PCA when:

Many features

Features are correlated

Visualization needed

Before clustering

Speed matters"""
```

```python
"""PCA NEEDS SCALING
PCA is sensitive to the scale of the data. If features are on different scales, PCA may give more weight to those with larger ranges, leading to skewed results. Therefore, it is important to standardize or normalize the data before applying PCA to ensure that all features contribute equally to the analysis.
"""
```

```python
import pandas as pd


df = pd.read_csv(r"C:\Users\USER\Downloads\Mall_Customers.csv")

X = df[["Annual Income (k$)", "Spending Score (1-100)"]]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Projection of Customers")
plt.show()
```

```python
X_full = df[["Age", "Annual Income (k$)", "Spending Score (1-100)"]]

X_scaled = scaler.fit_transform(X_full)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

pca.explained_variance_ratio_
```
