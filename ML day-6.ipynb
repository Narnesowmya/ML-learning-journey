```python
# Pandas
"""Pandas is a Python library used for working with data sets.

It has functions for analyzing, cleaning, exploring, and manipulating data.
It is built on top of another package called NumPy, which provides support for multi-dimensional arrays.
Pandas makes data analysis and manipulation easy and intuitive.
Pandas introduces two new data structures to Python: Series and DataFrame.
A Series is a one-dimensional array-like object that can hold many data types, including integers, strings, and floating-point numbers.
A DataFrame is a two-dimensional table made up of a collection of Series.
Pandas is widely used in data science, machine learning, and scientific computing.
"""
```

```python
import pandas   #importing pandas package

mydataset = {
  'cars': ["BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2]
}

myvar = pandas.DataFrame(mydataset)#create a DataFrame

print(myvar)
```

```python
import pandas as pd

a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
```

```python
#What is a DataFrame?
"""A DataFrame is a two-dimensional data structure, like a table in a database, an Excel spreadsheet, or a data frame in R.
It is a widely used data structure in data analysis and manipulation, as it allows for easy handling of large datasets with multiple rows and columns.
DataFrames can be created from various data sources, such as CSV files, Excel files, SQL databases, and more.
DataFrames provide a variety of functions and methods for data manipulation, such as filtering, sorting, grouping, and aggregating data.
DataFrames also support missing data handling, data alignment, and merging/joining of multiple datasets.
Pandas DataFrames are built on top of NumPy arrays, which provide efficient storage and manipulation of numerical data."""
```

```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df) 
```

```python
print(df.loc[0])
```

```python
#use a list of indexes:
print(df.loc[[0, 1]])
```

```python
"""A simple way to store big data sets is to use CSV files (comma separated files).

CSV files contains plain text and is a well know format that can be read by everyone including Pandas.

we will be using a CSV file called 'data.csv'."""
```

```python
import pandas as pd

df = pd.read_csv('data.csv')

print(df.to_string()) 
```

```python
"""Big data sets are often stored, or extracted as JSON.

JSON is plain text, but has the format of an object, and is well known in the world of programming, including Pandas we will be using a JSON file called 'data.json'.

Open data.json."""
```

```python
import pandas as pd

df = pd.read_json('data.json')

print(df.to_string()) 
```
