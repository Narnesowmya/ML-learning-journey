```python
# Operators in python
""" operators are  used to perform operation on variables and values.
Operators are : special symbols like +,-,/etc..
Operand : Values on which operators perform operations are called operands.
"""
```

```python
#Operator are of following types:
#1.Arithematic operators : +,-,*,/,%,**,//
a = 26
b = 25
print("Arithmetic Operators:")
print("Addition:",a+b)#a,b are operand and + is a operator,o/p:51
print("Subtraction:",a-b)#o/p:1
print("multiplicaton:",a*b)#o/p;650
print("Division:",a/b)#o/p:1.04
print("Mudolous:",a%b)#o/p:1
print("Exponent:",a**b)#o/p:26 raised to power 25
print("Floor Division:",a//b)#o/p:1

```

```python
#2.Assignment Operatos : =,==,+=,-=,*=,/=,%=,//=,**=
print("\nAssignment Operators:")
c = 2
c += 3 #c = c + 3
print(" c+=3:",c)#o/p:5
c -= 1 # c = c - 1
print("c-=1:",c)#o/p:4
c *= 2 # c = c * 2
print("c*2:",c)#o/p:8
c /= 4 # c = c / 4
print("c/=4:",c)#o/p:2.0
c **= 3 # c= c ** 3
print("c**=3:",c)#o/p:8..0
c //=2 # c = c // 2
print("c//=2:",c)#o/p:4.0
c %= 3 # c = c % 3 
print("c %=3:",c)#o/p:1,0
```

```python
#3.Comparison Operator : ==,!=,>,<,>=,<=
print("\nComparison Operators:")
x = 15 
y = 20
print("x == y:",x==y)#o/p:False
print("x != y:",x!=y)#o/p"True
print("x > y:",x>y)#o/p:False
print("x < y:",x<y)#o/p:True
print("x >= y:",x>=y)#o/p:False
print("x <= y:",x<=y)#o/p:True
```

```python
#4.Logical Operators : and,or,not
print("\nLogical Operators:")
p = True
q = False
print(" p and q:",p and q)#o/p:False
print(" p or q:", p or q)#o/p:True
print("not p:",not p )#o/p:False
print("not q:",not q)#o/p:True
```

```python
#5.Bitwise Operators : &,|,^,~,<<,>>
print("\nBitwise Operatos:")
m = 24 
n = 12
print("m & nA:",m & n)#o/p:8
print(" m | n:",m | n)#o/p:28
print( m ^ n)#o/p:20
print("~m:",~m)#o/p:25
print("m << 2 :",m << 2)#o/p:96
print("n >> 2:",n>>2)#o/p:3
```

```python
#6.Identity Operators : is, is not
print("\nIdentity Operators:")
a = 10
b = 10
c = 20
print("a is b:", a is b)  # o/p: True
print("a is c:", a is c)  # o/p: False
print("a is not c:", a is not c)  # o/p: True
```
