```python
#Mean means average of a data set or numbers.
#Mean = Sum of all numbers / Total count of numbers
data = [12, 15, 14, 10, 8, 20, 16]
mean = sum(data) / len(data)
print("Mean:", mean)
```

```python
# mean in Machine Learning is used to summarize a dataset with a single value that represents the central tendency of the data. It is commonly used in various applications such as:
# 1. Data Preprocessing: Mean is used to fill in missing values in datasets (mean imputation).
# 2. Feature Scaling: Mean is used in normalization techniques to scale features.   
# 3. Model Evaluation: Mean is used to calculate metrics like Mean Squared Error (MSE) to evaluate model performance.
# 4. Clustering Algorithms: Mean is used in algorithms like K-Means to determine the centroids of clusters.
#Overall, mean is a fundamental statistical measure that helps in understanding and analyzing data in machine learning tasks.
```

```python
data = [10, 20, 30, 40]
mean = (10 + 20 + 30 + 40) / 4
print("Mean:", mean)
```

```python
#Median is the middle value in a data set when the numbers are arranged in order. To find the median, you first sort the numbers and then identify the middle number. If there is
# an even number of values, the median is the average of the two middle numbers.
data = [12, 15, 14, 10, 8, 20, 16]
sorted_data = sorted(data)# Sort the data
n = len(sorted_data)
if n % 2 == 1:
    median = sorted_data[n // 2]# Middle number
else:
    median = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2# Average of the two middle numbers
print("Median:", median)
```

```python
#Median in Machine Learning is used to summarize a dataset by identifying the central value, which is particularly useful in the following scenarios:
#1. Handling Skewed Data: Median is less affected by outliers and skewed distributions, making it a better measure of central tendency in such cases.
#2. Data Preprocessing: Median can be used to fill in missing values (median imputation) in datasets.
#3. Robust Statistics: Median is often used in robust statistical methods that aim to reduce the influence of outliers.
#4. Feature Engineering: Median can be used to create new features that capture the central tendency of a variable.
#Overall, median is a valuable statistical measure in machine learning for understanding data distributions and improving model performance.
```

```python
data = [10, 20, 30, 40, 1000]# Skewed data with an outlier
sorted_data = sorted(data)  # Sort the data
n = len(sorted_data)        
if n % 2 == 1:
    median = sorted_data[n // 2]  # Middle number
else:
    median = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2  # Average of the two middle numbers
print("Median with outlier:", median)  # The median is less affected by the outlier.
""" Statistics for Machine Learning  basics Day 8 - 10   day - 8 : Mean , Median.
"""
```

```python
#DAY 9 — VARIANCE & STANDARD DEVIATION
#Variance is a measure of how much the numbers in a data set vary or spread out from the mean (average) of that data set. It quantifies the average squared difference between each number and the mean.
data = [12, 15, 14, 10, 8, 20, 16]
mean = sum(data) / len(data)
squared_differences = [(x - mean) ** 2 for x in data]
variance = sum(squared_differences) / len(data)
print("Variance:", variance)
#Variance in Machine Learning is used to measure the spread of data points around the mean. It helps in understanding the distribution of data and is crucial for various applications such as:
#1. Model Evaluation: Variance is used to assess the performance of machine learning models, particularly in regression tasks.
#2. Bias-Variance Tradeoff: Variance is a key component in understanding the bias
# and variance tradeoff, which is essential for model selection and tuning.
#3. Feature Selection: Variance can be used to identify important features in a dataset, as features with low variance may not contribute significantly to the model.
#Overall, variance is a fundamental statistical measure that aids in analyzing data and improving machine learning models.
```

```python
data = [10, 20, 30, 40]
mean = sum(data) / len(data)
squared_differences = [(x - mean) ** 2 for x in data]
variance = sum(squared_differences) / len(data)
print("Variance:", variance)
data.append(1000)
squared_differences = [(x - mean) ** 2 for x in data]
variance = sum(squared_differences) / len(data)
print("Variance after adding an outlier:", variance)  # Variance is sensitive to outliers.
```

```python
#satandard Deviation is a measure of the amount of variation or dispersion in a set of values. It is the square root of the variance and provides a measure of how spread out the numbers are around the mean.
data = [12, 15, 14, 10, 8, 20, 16]
mean = sum(data) / len(data)
squared_differences = [(x - mean) ** 2 for x in data] # x means each value in data, mean is the average value of data.
variance = sum(squared_differences) / len(data)
standard_deviation = variance ** 0.5
print("Standard Deviation:", standard_deviation)
#Standard Deviation in Machine Learning is used to quantify the amount of variation or dispersion in a dataset. It is important for several reasons:
#1. Data Preprocessing: Standard deviation is used in normalization techniques to scale features, ensuring that they have a mean of 0 and a standard deviation of 1.
#2. Model Evaluation: Standard deviation is used to assess the performance of machine learning models, particularly in regression tasks.
#3. Outlier Detection: Standard deviation helps in identifying outliers in the data, as values that are several standard deviations away from the mean may be considered outliers.
#Overall, standard deviation is a crucial statistical measure that aids in understanding data distributions and improving machine learning models.
```
